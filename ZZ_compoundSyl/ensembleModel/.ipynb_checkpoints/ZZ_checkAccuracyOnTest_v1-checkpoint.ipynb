```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import r2_score,make_scorer,mean_squared_error
from sklearn.model_selection import GroupKFold
from scipy.stats import wilcoxon
from scipy import signal
import numpy as np
import os
from sklearn.base import clone
from matplotlib import pyplot as plt
```

```python
model_path = '/mnt/z4/zz367/EphysMONAO/Analyzed/Figures/pair5CU29CU55/CompoundSyl/pair5RigCCU29/ensembleModel/XGB_res/20240917-ch13/Call_Acoustic/Call_Acoustic.model.fold0.json'
fn_test_predictor = '/mnt/z4/zz367/EphysMONAO/Analyzed/Figures/pair5CU29CU55/CompoundSyl/pair5RigCCU29/ensembleModel/XGB_res/20240917-ch13/Compound_Acoustic/Compound_Acoustic.predictor.csv'
fn_test_output = '/mnt/z4/zz367/EphysMONAO/Analyzed/Figures/pair5CU29CU55/CompoundSyl/pair5RigCCU29/ensembleModel/XGB_res/20240917-ch13/Compound_Acoustic/Compound_Acoustic.output.csv'
fn_save = '/mnt/z4/zz367/EphysMONAO/Analyzed/Figures/pair5CU29CU55/CompoundSyl/pair5RigCCU29/ensembleModel/XGB_res/20240917-ch13/General_call_compound/model0.csv'
```

```python
# model = clone(model_base)
model = xgb.XGBRegressor()
model.load_model(model_path)
```

```python
## read data
X_te = np.loadtxt(fn_test_predictor, delimiter=',')
y_te = np.loadtxt(fn_test_output, delimiter=',')
print(X_te.shape, y_te.shape)
```

```python
# null = train-fold mean rate
y_mean = np.mean(y_te)
y_null = np.full_like(y_te, y_mean, dtype=float)

# ===== fit model =====
y_hat = np.clip(model.predict(X_te), 1e-9, None)

# save metrics
r2s_model = r2_score(y_te, y_hat)
s_model = -mean_squared_error(y_te, y_hat)
s_null  = -mean_squared_error(y_te, y_null)
deltas = s_model - s_null
pct_improvements_full = 100.0 * (s_null - s_model) / s_null
```

```python
# save metrics
metric = [r2s_model, s_model, s_null, deltas, pct_improvements_full]
fd_save = os.path.dirname(fn_save)
if not os.path.exists(fd_save):
    os.makedirs(fd_save)
np.savetxt(fn_save, metric)
```
