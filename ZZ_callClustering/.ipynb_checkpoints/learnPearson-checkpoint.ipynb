## 2. Make spectrograms

```python
# define parameters for VAE and spectrograms
X_SHAPE = [128, 128]

p = {
    'get_spec': vae_goffinet.get_specZZ, # spectrogram maker
    'max_dur': 1e9, # maximum syllable duration
    'min_freq': 400, # minimum frequency
    'max_freq': 10000, # maximum frequency
    'num_freq_bins': X_SHAPE[0], # hard-coded
    'num_time_bins': X_SHAPE[1], # hard-coded
    'nperseg': 512, # FFT
    'noverlap': 256, # FFT, determines window overlap when calculating spectrograms
    'spec_min_val': 2, # minimum log-spectrogram value
    'spec_max_val': 10, # maximum log-spectrogram value
    'fs': 44100, # audio samplerate
    'mel': True, # frequency spacing, mel or linear
    'time_stretch': True, # stretch short syllables?
    'within_syll_normalize': False, # normalize spectrogram values on a # spectrogram-by-spectrogram basis
    'max_num_syllables': None, # maximum number of syllables per directory
    'sylls_per_file': 20, # syllable per file
    'real_preprocess_params': ('min_freq', 'max_freq', 'spec_min_val', 'spec_max_val'), # tunable parameters
    'int_preprocess_params': ('nperseg','noverlap'), # tunable parameters
    'binary_preprocess_params': ('time_stretch', 'mel', 'within_syll_normalize'), # tunable parameters
    'window_length': 0.032,  # for continuous sliding: size of the spectrogram window, unit is sec
    'hop_length': 0.004  # for continuous sliding: how much to hop for successive spectrogram windows, unit is sec
}
```

```python
win_ms = p['nperseg']/p['fs']*1000
hop_ms = (p['nperseg']-p['noverlap'])/p['fs']*1000
print(f'Spectrogram window size (ms): {win_ms:.3f}')
print(f'Spectrogram window hop (ms): {hop_ms:.3f}')
```

```python
# grab all wav files
fns_wav = sorted(glob.glob(os.path.join(fd_data, birdID, '*.wav')))
len(fns_wav)
```

```python
# loop through files determine the max syllable duration
fns_label = sorted(glob.glob(os.path.join(fd_data, birdID, '*.label.txt')))
def ZZ_getDur_v1(fn_label, fs):
    dur = []
    labels = np.genfromtxt(fn_label, dtype=str, delimiter=None, encoding='utf-8-sig')
    # check if it has target syllable
    idx = [ii for ii in range(labels.shape[0]) if labels[ii] in syl]
    if len(idx)>0:
        fn_time = fn_label.replace('.label.txt', '.time.txt')
        seg = np.loadtxt(fn_time, delimiter=',', dtype=int)
        dur = list((seg[idx,1] - seg[idx,0]) / fs);
    return(dur)

# loop through all wav files
with Parallel(n_jobs=48, verbose=5) as parallel:
    dur_all = parallel(delayed(ZZ_getDur_v1)(fn, fs=p['fs']) for fn in fns_label)
    
dur_flat = [item for sublist in dur_all for item in sublist]
max_dur = max(dur_flat)
print(max_dur)
```

```python
fi = 1
fn = fns_wav[fi]
print(fn)
```

```python
def ZZ_specFromWavGoffinet_v1(fn, p, syl, max_dur): 
    # calculate spectrograms from wav file
    # check if annotation file exist
    fn_label = fn.replace('.wav','.label.txt')
    # given wav file and selected syllables, output spectrograms and meta info
    specs = []
    info = pd.DataFrame()
    # when calcualte spectrogram, include a small padding to avoid artifacts at the boundary
    if os.path.exists(fn_label):
        labels = np.genfromtxt(fn_label, dtype=str, delimiter=None, encoding='utf-8-sig')
        # check if it has target syllable
        idx = [ii for ii in range(labels.shape[0]) if labels[ii] in syl]
        if len(idx)>0:
            fn_time = fn_label.replace('.label.txt', '.time.txt')
            seg = np.loadtxt(fn_time, delimiter=',', dtype=int)
            # load wav file
            fs, audio = wavfile.read(fn)
            duration = len(audio)/fs
            for m in idx:
                # determine onset and offset of syllables
                onset = seg[m,0] / fs
                offset = seg[m,1] / fs
                spec, flag, target_freqs, target_times = vae_goffinet.get_specZZ(onset, offset, audio, p, fs=fs, target_times=None, remove_dc_offset=True, max_dur=max_dur)

                specs.append(spec)
                # record the meta info of this syllable
                row = pd.DataFrame([{'fn_wav':fn, 'istart':seg[m,0], 'iend':seg[m,1], 'label':labels[m]}])
                info = pd.concat([info, row], ignore_index=True)
    return(specs, info)
```

```python
# loop through all wav files
with Parallel(n_jobs=48, verbose=5) as parallel:
    res = parallel(delayed(ZZ_specFromWavGoffinet_v1)(fn, p, syl, max_dur) for fn in fns_wav)
```

```python
# flatten the result
temp = [aa[0] for aa in res]
specs = [arr for sublist in temp if sublist for arr in sublist]
df_list = [aa[1] for aa in res]
info = pd.concat([df for df in df_list if not df.empty], ignore_index=True)
print(len(specs), info.shape)
```

### Obtain latent representation

```python
train_dataloader = DataLoader(train_data, batch_size=64, shuffle=False, num_workers=4)  # set shuffle to false to match the order in id_pd
```

```python
info.shape
```

```python
# loop through dataloader, obtain model latent space
latent_m = np.zeros((info.shape[0], 32))
latent_d = np.zeros((info.shape[0], 32))
recon = np.zeros((info.shape[0], X_SHAPE[0], X_SHAPE[1]))
model.eval()
count = 0
for i, data in tqdm.tqdm(enumerate(train_dataloader)):
    data = data.to('cuda:0')
    with torch.no_grad():
        _, _, rec, mu, d = model.forwardZZ(data, return_latent_rec=True)
        a = rec.shape[0]
        latent_m[count:(count+a),:] = mu
        latent_d[count:(count+a),:] = d
        recon[count:(count+a),:,:] = rec
        count += a
```
